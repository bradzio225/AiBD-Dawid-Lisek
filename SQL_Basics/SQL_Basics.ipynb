***Zadanie 11***

Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.


```python
usa_rented_categories = pd.read_sql("""select distinct (name) from country, city, address, customer, rental, inventory,
                    film_category, category where
                    (country.country in ('United States')) and
                    (country.country_id = city.country_id) and
                    (city.city_id = address.city_id) and
                    (address.address_id = customer.address_id) and
                    (customer.customer_id = rental.customer_id) and
                    (rental.inventory_id = inventory.inventory_id) and
                    (inventory.film_id = film_category.film_id) and
                    (film_category.category_id = category.category_id)""", con=connection_sqlalchemy)
usa_rented_categories
```
