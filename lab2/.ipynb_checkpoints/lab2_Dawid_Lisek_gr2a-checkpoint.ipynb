```python
#importowanie bibliotek
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
#otwarcie oryginalnego pliku
with open('Original_Data/weather.txt', "r") as f:
    weather_original = f.read()

# usuwanie separatorów 'I' oraz 'S'
weather_cleaned = re.sub(r"[OSD]|I(?![A-Z])", "", weather_original)
# usunięcie spacji
weather_cleaned = re.sub(r" +", " ", weather_cleaned)
# usunięcie połączonych napisów z danymi
weather_cleaned = re.sub(r"(?<=[A-Z])\-", " -", weather_cleaned)
```

```python
# stworzenie nowego pliku .txt z oczyszczonymi danymi
with open('Analysis_Data/weather_cleaned.txt', 'w') as f:
    f.write(weather_cleaned)
f.close()
```

```python
#wczytanie plików jako dataframe
weather = pd.read_csv('Analysis_Data/weather_cleaned.txt', sep = ' ', header = None) 
weather
```

```python
#
weather['year'] = weather[0].str.slice(start=11, stop=15)
weather['month'] = weather[0].str.slice(start=15, stop=17)
weather['element'] = weather[0].str.slice(start=17)
weather[0] = weather[0].str.slice(start = 0, stop = 11)


dict_of_days = {}
for days in range(1, 32):
    dict_of_days[days] = f'Day{days}'
weather.rename(columns={0: 'id'}, inplace=True)
weather.rename(columns=dict_of_days, inplace=True)
weather.drop(32, axis=1, inplace=True)
weather = weather[[weather.columns[0], *weather.columns[32:35], *weather.columns[1:32]]]
# zamiana wartości na wartości liczbowe
weather.iloc[:, 4:] = weather.iloc[:, 4:].apply(pd.to_numeric, errors='coerce')
weather.replace(-9999, np.NaN, inplace=True)
weather.iloc[:, 4:] = np.divide(weather.iloc[:, 4:], 10)
weather
```

```python
# przegrupowanie danych 'rok' 'miesiąc' 'element w zależności od 'dnia'
melt_weather = pd.melt(weather, id_vars=['year', 'month', 'element'], var_name='day', value_vars=[*dict_of_days.values()])
melt_weather
```

```python
# usunięcie elementów 'NaN'
melt_weather.dropna(inplace=True)
# usunięcie wartości zerowych '0'
melt_weather = melt_weather[melt_weather.value != 0]
melt_weather = melt_weather[(melt_weather.year > '2009') & (melt_weather.year < '2011')]
melt_weather_2010 = melt_weather.reset_index(drop=True)
melt_weather_2010.to_csv('Analysis_Data/melt_weather.csv')
```

```python
# połączenie roku, miesiąca i dnia w jedną kolumnę
melt_weather_2010['Date'] = melt_weather_2010['year'] + "-" + melt_weather_2010['month'] + "-" + melt_weather_2010['day'].str.slice(start=3)
weather_pivot = melt_weather_2010.pivot_table(index='Date', columns='element', values='value')
weather_pivot.drop(columns='PRCP', inplace=True)
weather_pivot.dropna(inplace=True)
weather_pivot
```

```python
T_MAX = []
T_MIN = []
months_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
for month in range(1, 10):
    T_MAX.append(weather_pivot.filter(like=f'2010-0{month}', axis=0)['TMAX'].max())
    T_MIN.append(weather_pivot.filter(like=f'2010-0{month}', axis=0)['TMIN'].min())
for month in range(10, 13):
    T_MAX.append(weather_pivot.filter(like=f'2010-{month}', axis=0)['TMAX'].max())
    T_MIN.append(weather_pivot.filter(like=f'2010-{month}', axis=0)['TMIN'].min())
```

```python
print(T_MAX)
print(T_MIN)
```

```python
w = 0.4
br1 = np.arange(len(months_list))
br2 = [x + w for x in br1]
plt.figure(figsize=(15,6))
plt.bar(br1, T_MAX, w,  color='red', label="T_MAX")
plt.bar(br2, T_MIN, w, color='blue', label='T_MIN')
plt.xticks(br1+w/2, months_list)
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.title("Wykres maksymalnej i minimalnej temperatury w 2010 roku")
plt.legend()
plt.show()
```
