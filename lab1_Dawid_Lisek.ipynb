```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1
f = lambda x: x**2 + 5
```

1. $x>-1$ oraz $x<1$

```python
#1
przedzial1 = np.linspace(-1, 1)
plt.plot(przedzial1, f(przedzial1), label = "x > -1 & x < 1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Wykres funkcji dla argumentów x <-1, 1>")
plt.show()
```

2. $x>-6$ oraz $x<6$

```python
#2
przedzial2 = np.linspace(-6, 6)
plt.plot(przedzial2, f(przedzial2), label = "x > -6 & x < 6")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Wykres funkcji dla argumentów x <-6, 6>")
plt.show()
```

```python
#3
przedzial3 = np.linspace(0, 5)
plt.plot(przedzial3, f(przedzial3), label = "x > 0 & x < 5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Wykres funkcji dla argumentów x <0, 5>")
plt.show()
```
